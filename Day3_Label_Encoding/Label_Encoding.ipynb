## Label Encoding

## Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### Label Encoding of Breast Cancer Dataset

```python
# loading the data from csv file to pandas dataframe
cancer_data = pd.read_csv("E:\ML_Series\ML\Day3_Label_Encoding\Data_Sets\data.csv")
```

```python
# first 5 rows of the dataframe
cancer_data.head()
```

```python
# finding the count of different diagnosis label
cancer_data['diagnosis'].value_counts()
```

```python
# load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
print(labels)
```

```python
# appending the laels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

```python
cancer_data['target'].value_counts()
```
